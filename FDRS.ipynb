```python
# Results in page 16 Fig3 different fractional orders and tau = 4.02
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def fractional_derivative(x, alpha, dt):
    N = len(x)
    derivative = np.zeros(N)
    for i in range(N):
        sum_term = 0
        for j in range(i):
            sum_term += x[j] * (i - j)**(-alpha)
        derivative[i] = (x[i] - sum_term) / dt**alpha
    return derivative

# Parameters
sigma = 10.0
r = 1.2
b = 8.0 / 3.0
tau = 4.13
alpha_x = 0.91
alpha_y = 0.94
alpha_z = 0.96
t_max = 1200.0  
h = 0.05  

# Initial conditions
x0, y0, z0 = 0.5, 0.45, 0.1

# Number of time steps
num_steps = int(t_max / h)

# Arrays to store results
t_values = np.linspace(0, t_max, num_steps)
x_values = np.zeros(num_steps)
y_values = np.zeros(num_steps)
z_values = np.zeros(num_steps)

# Precompute fractional derivatives
x_derivative = fractional_derivative(np.zeros(num_steps), alpha_x, h)
y_derivative = fractional_derivative(np.zeros(num_steps), alpha_y, h)
z_derivative = fractional_derivative(np.zeros(num_steps), alpha_z, h)

# Initial values
x_values[0] = x0
y_values[0] = y0
z_values[0] = z0

# Simulate the fractional delay Lorentz system 
for i in range(1, num_steps):
    x_values[i] = x_values[i-1] + h * (sigma * (y_values[i-int(tau/h)] - x_values[i-1]) + x_derivative[i])
    y_values[i] = y_values[i-1] + h * (x_values[i-1] * (r - z_values[i-1]) - y_values[i-int(tau/h)] + y_derivative[i])
    z_values[i] = z_values[i-1] + h * (x_values[i-1] * y_values[i-int(tau/h)] - b * z_values[i-1] + z_derivative[i])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t_values, x_values, label='x(t)')
plt.xlabel('Time')
plt.ylabel('Values')
plt.title('Fractional Delay Lorentz System')
plt.legend()
plt.show()
```
